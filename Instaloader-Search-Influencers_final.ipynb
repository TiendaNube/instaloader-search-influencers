```python
import instaloader
import numpy as np

# Create an instance of Instaloader class
bot = instaloader.Instaloader()

# Interactive login on terminal
bot.interactive_login('your_instagram_profile') # Asks for password in the terminal
```

```python
search_similar_profiles = ['emprender.mejor', 'emprendersimple', 'vickybecci', 'ikigai.contenidos', 'digitalizadas_', 
                           'marketingconbelu', 'mamita.power.emprende'
                          ]
```

```python
# Load a profile from an Instagram handle

similar_accounts = []

for user_insta in search_similar_profiles:
    
    profile = instaloader.Profile.from_username(bot.context, user_insta) 
    
    for a in profile.get_similar_accounts():
        similar_accounts.append(a.username)
    
```

```python
similar_accounts = list(np.unique(np.array(similar_accounts)))
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd

from pytube import YouTube
from pytube import Channel
        
import requests
import bs4
from bs4 import BeautifulSoup
import csv
import re
import time
from datetime import timedelta
from datetime import datetime
from datetime import date

from itertools import dropwhile, takewhile
from pytz import timezone


SINCE = datetime.today() + timedelta(days=-90)  # further from today, inclusive 
UNTIL = datetime.today() + timedelta(days=0)  # closer to today, not inclusive
```

```python

headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'}

#---------


file = open('Backlog_Influencers_AR.txt','a+')

for USER_PROFILE in similar_accounts:
    
    if USER_PROFILE not in file:
    
        profile = instaloader.Profile.from_username(bot.context, USER_PROFILE)

        full_name = profile.full_name
        if full_name is None: 
            full_name = '-'

        followers = profile.followers
        if followers is None: 
            followers = '-'

        followees = profile.followees
        if followees is None: 
            followees = '-'
            
        external_url = profile.external_url
        youtube='-'
        yt_avg_views='-'
        if external_url is None: 
            external_url = '-'

            
        business_category_name = profile.business_category_name
        if business_category_name is None:
            business_category_name = '-'

        is_business_account = profile.is_business_account

        is_verified = profile.is_verified      

        igtvcount = profile.igtvcount
        if igtvcount is None: 
            igtvcount = '-'

        aux_biography = profile.biography
        aux_biography = aux_biography.lower()
        biography = '-'
        engagement_insta = '-'
        if ('commerce' in aux_biography) or ('dropshipping' in aux_biography) or ('marketing' in aux_biography) or ('ecom' in aux_biography) or ('tiendanube' in aux_biography) or ('mercadoshop' in aux_biography) or ('empretienda' in aux_biography) or ('shopify' in aux_biography) or ('emprender' in aux_biography) or ('emprendimiento' in aux_biography) or ('plataforma' in aux_biography) or ('cursos online' in aux_biography) or ('especialistas ecommerce' in aux_biography) or ('tiendas online' in aux_biography) or ('ayudo a crecer y vender' in aux_biography) or ('trafico' in aux_biography) or ('negocio online' in aux_biography) or ('negocios digitales' in aux_biography) or ('potenciamos tu negocio' in aux_biography) or ('potenciar' in aux_biography) or ('loja virtual' in aux_biography) or ('tienda virtual' in aux_biography) or ('mkt' in aux_biography)  or ('alumnos' in aux_biography) or ('ventas' in aux_biography):
            if ('investimento' not in aux_biography) and ('renda' not in aux_biography) and ('investidor' not in aux_biography) and ('trader' not in aux_biography) and ('crossfit' not in aux_biography) and ('cripto' not in aux_biography) and ('concurso' not in aux_biography) and ('rendimento' not in aux_biography):
                biography = 'has_key_word'

                # Engagement rate of 90 last days
                num_followers = profile.followers
                total_num_likes = 0
                total_num_comments = 0
                total_num_posts = 0
                count = 0
                
                k = 0  # initiate k

                for post in profile.get_posts():
                    postdate = post.date

                    if postdate > UNTIL:
                        continue
                    elif postdate <= SINCE:
                        k += 1
                        if k == 100:
                            break
                        else:
                            continue
                    else:
                        k = 0  # set k to 0
                        total_num_likes += post.likes
                        total_num_comments += post.comments
                        total_num_posts += 1
                        count +=1
                try:
                    engagement_insta = float(total_num_likes/total_num_posts + total_num_comments/total_num_posts) / (num_followers)
                except:
                    engagement_insta = '-'
            try:
                #YouTube: 
                url = external_url
                data = requests.get(url)
                html = data.text
                soup = BeautifulSoup(html,"html.parser")  
                youtube_aux = []
                for n in soup.findAll("a", href=re.compile('youtube.com')): 
                    if n.get('href') not in youtube:
                        youtube_aux.append(n.get('href'))
                        youtube = n.get('href').replace("https://www.youtube.com/howyoutubeworks?", "").replace("?view_as=subscriber&ltclid=", "").replace("?utm_source=instagram&utm_medium=linktree&utm_campaign=youtube&ltclid=", "").replace("https://www.youtube.com/howyoutubeworks?utm_campaign", "").replace("/?ltclid=", "/")

                        total_num_views = 0
                        total_num_videos = 0
                        count = 0
                        videos = []
                        videos_aux = []
                        videos_aux = (youtube + "/videos").replace("//", "/")
                        videos.append(videos_aux)
                        for video in videos: 
                            channel_info = Channel(video) 
                            for url in channel_info.url_generator(): 
                                video_details = YouTube(url) 
                                postdate = video_details.publish_date
                                if postdate >= SINCE:
                                    if postdate <= UNTIL:
                                        total_num_views += video_details.views 
                                        total_num_videos += 1
                                        count +=1
                                else: break

                        yt_avg_views = float(total_num_views) / (total_num_videos)
            except:
                youtube='-'
                yt_avg_views='-'
                
        print(USER_PROFILE)
        
        file.write(USER_PROFILE + ';;' + full_name + ';;' + str(followers) + ';;' + str(followees) 
                    + ';;' + str(igtvcount)
                    + ';;' + str(engagement_insta)
                    + ';;' + biography  
                    + ';;' + external_url
                    + ';;' + youtube
                    + ';;' + str(yt_avg_views)
                    + ';;' + str(is_verified)
                    + ';;' + str(is_business_account) + ';;' + business_category_name  + "\n")
        file.flush()


file.close()
```
